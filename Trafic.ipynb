```python
import pandas as pd
df = pd.read_csv("Traffic-Prediction-using-Machine-Learning.csv")
df
```

## 1. What kind of preprocessing was done to the dataset before trainning the model ##

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
le = LabelEncoder()
df['Day_Encoder'] = le.fit_transform(df['Day'])
print(df)
```

```python
df
```

```python
df.info()
```

```python
Feature = df[['Weather','Temperature','CodedDay','Zone','Weather','Temperature','Day_Encoder']]
Feature
```

```python
target=df[['Traffic']]
target
```

## 2. Which machinglearning algorithm was useed to predict traffic,and why was it chosen ##

```python
print(df.isnull().sum())
```

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(Feature,target,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dect = DecisionTreeClassifier()
dect.fit(x_train,y_train)
```

```python
dect.score(x_test,y_test)
```

## 4.What metric was used to evaluate the model performance, and how is it calculated? ##

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy= accuracy_score(y_test,y_pred)
```

```python
print(f"Model Accuracy:{accuracy:.2f}")
```

```python
error_rate=1 - accuracy
print("Error_rate:",error_rate)
```
