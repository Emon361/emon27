## 4.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Features = df[['Order_Quantity','Profit','Revenue']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(Features)
```

```python
inertia=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,9))
plt.plot(range(1,11),inertia,marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.tight_layout()
plt.show()
           
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters= optimal_k,random_state=42)
df['Cluster']=kmeans.fit_predict(scaled_features)
```

```python
sns.pairplot(df[['Order_Quantity','Profit','Revenue','Cluster']],hue='Cluster',palette='Set2')
plt.suptitle('Customer Clusters based on Purchasing behavior',y=1.02)
plt.show()
             
```

## 11.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df['High_Order']=df['Order_Quantity']>=10
```

```python
features =['Customer_Age','Customer_Gender','Age_Group','Product_Category','Sub_Category']
X=df[features]
Y=df['High_Order']
```

```python
X_encoded =X.copy()
le = LabelEncoder()
for col in ['Customer_Age','Customer_Gender','Age_Group','Product_Category','Sub_Category']:
    X_encoded[col] = le.fit_transform(X[col])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_encoded,Y,test_size=0.3,random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=100,random_state=42)
clf.fit(X_train,Y_train)
```

```python
Y_pred = clf.predict(X_test)
print('Confusion Matrix:\n',confusion_matrix(Y_test,Y_pred))
print('\nClassification Report:\n',classification_report(Y_test,Y_pred))
```

## 12.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
features =['Customer_Age','Customer_Gender','Age_Group','Product_Category','Sub_Category','Unit_Cost','Unit_Price']
X=df[features]
Y=df['Profit']
```

```python
X_encoded =X.copy()
le = LabelEncoder()
for col in ['Customer_Gender','Age_Group','Product_Category','Sub_Category']:
    X_encoded[col] = le.fit_transform(X[col])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_encoded,Y,test_size=0.3,random_state=42)
```

```python
reg = RandomForestRegressor(n_estimators=100,random_state=42)
reg.fit(X_train,Y_train)
```

```python
Y_pred = reg.predict(X_test)
mse = mean_squared_error(Y_test,Y_pred)
r2=r2_score(Y_test,Y_pred)
print(f"Mean Squared Error : {mse:.2f}")
print(f"R^2 Score:{r2: .2f}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
features = df[['Order_Quantity','Profit','Revenue']]
```

```python
sclaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters = 4, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
```

```python
df['Cluster'] = clusters
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)
df['PCA1'] =pca_result[:,0]
df['PCA2'] = pca_result[:, 1]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data = df , x='PCA1',y='PCA2',hue='Cluster',palette='tab10',alpha=0.6)
plt.title('Customer Segments Based on Purchasing Behavior')
plt.xlabel('Principle Component 1')
plt.ylabel('Principle Component 2')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```
